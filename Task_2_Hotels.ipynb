```python
import pandas as pd


df = pd.read_excel('Riyadh_Hotels.xlsx')


print("--- عرض عينة من البيانات ---")
display(df.head())

print("\n--- فحص القيم المفقودة ---")
print(df.isnull().sum())


print("\n--- عدد الصفوف المكررة ---")
print(df.duplicated().sum())
```

الصفوف المكررة (Duplicates):تم رصد (٢٢) صفوف مكررة ويجب حذفها لتجنب تكرار النتايج

نوع البيانات (Data Types):عمود السعر سليم تقنياً ومخزن كأرقام عشرية (float64)، مما يسمح لنا بإجراء العمليات الإحصائية مباشرة.

ومع ذلك، يوجد 60 قيمة مفقودة في عمود price و 249 في base_price يجب أخذها في الاعتبار عند حساب المتوسطات.
