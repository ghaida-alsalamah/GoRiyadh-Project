**Check for Missing Values**

```python
def check_missing_values(df):
  missing_values = df.isnull().sum()
  print("Missing Values per Column:")
  print(missing_values[missing_values > 0])
  if missing_values.any():
    print("Handle the missing values")
  else:
    print("\nNo Missing Values Found.")

check_missing_values(df)
```

```python
# Analyze missing values
missing_percentage = (df.isnull().sum() / len(df)) * 100
missing_data = pd.DataFrame({
    'Column': missing_percentage.index,
    'Missing_Percentage': missing_percentage.values
})
missing_data = missing_data[missing_data['Missing_Percentage'] > 0].sort_values('Missing_Percentage', ascending=False)

print("Missing Data Analysis:")
missing_data.head(10)
```

> `ratingSignals` (58.9% missing), `rating` (58.9% missing), `price` (7.8% missing), `likes` (0.01% missing)

**Check for Duplicate Values & Drop them**

```python
def check_duplicates(df):
  duplicates = df.duplicated().sum()
  print(f"Number of Duplicate Samples: {duplicates}")
  if duplicates > 0:
    print("Dropping Duplicates...")
    df.drop_duplicates(inplace=True)
    print("Duplicates Dropped.")
  else:
    print("No Duplicate Samples Found.")

check_duplicates(df)
```

# **Data CleaningüßΩ**

> copy the dataset for cleaning

```python
df_clean = df.copy()
```

> fill missing values with 'unknown'

```python
df_clean['price'] = df_clean['price'].fillna('unknown')
df_clean['likes'] = df_clean['likes'].fillna(df_clean['likes'].mode()[0])
df_clean['rating'] = df_clean['rating'].fillna(df_clean['rating'].mode()[0])
df_clean['ratingSignals'] =df_clean['ratingSignals'].fillna(df_clean['ratingSignals'].mode()[0])
check_missing_values(df_clean)
```

> drop cafes in the dataset

```python
df_clean = df_clean[~df_clean["categories"].str.lower().str.contains("cafe|coffee|caf√©|donut|dessert|ice cream", na=False)]
```

# **Plotsüìä**

```python
p1 = {'lat': 24.719462, 'lon': 46.719108}
df.loc[pd.isnull(df['price']), 'price'] = 'Unknown'
fig = px.scatter_mapbox(df,
                        lat=df['lat'],
                        lon=df['lng'],
                        center=p1,
                        color=df['price'],
                        zoom=9,
                        mapbox_style="open-street-map",
                        title='Riyadh Restaurants'
                       )
fig.show()
```

```python
p1 = {'lat': 24.719462, 'lon': 46.719108}
fig = px.density_mapbox(
    df_clean,
    lat="lat",
    lon="lng",
    z="likes",
    radius=15,
    center=p1,
    zoom=10,
    mapbox_style="open-street-map",
    title="Popular Spots by Likes"
)

fig.show()
```

```python
fig = px.scatter_mapbox(
    df_clean,
    lat="lat",
    lon="lng",
    size="likes",
    color="price",
    size_max=25,
    zoom=10,
    mapbox_style="open-street-map",
    hover_name="name",
    hover_data=["categories", "rating", "likes"],
    title="Popularity Map"
)

fig.show()
```

```python
fig = px.scatter(
    df_clean,
    x="rating",
    y="likes",
    color="price",
    size="likes",
    hover_name="name",
    hover_data=["categories"],
    log_y=True,
    title="Rating vs Popularity by Price"
)

fig.show()
```

```python
fig = px.pie(
    df_clean,
    names="price",
    hole=0.45,
    title="Restaurant Price Distribution in Riyadh"
)

fig.show()
```
