مراجعة وتحليل بيانات فنادق الرياض (Riyadh Hotels Data) Review

الهدف: فحص جودة البيانات وتحديد المشاكل والنقص قبل البدء بعمليات التحليل المتقدمة.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('Riyadh_Hotels.xlsx')

df.head()
```

```python

print("--- Summary Statistics ---")
display(df.describe())


missing_pct = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({'Column': missing_pct.index, 'Missing %': missing_pct.values})
print("\n--- Percentage of Missing Data ---")
display(missing_df.sort_values(by='Missing %', ascending=False))
```

```python

fig1 = px.pie(df, names='rating', title='توزيع التقييمات في فنادق الرياض (تفاعلي)', hole=0.3)
fig1.update_traces(textinfo='percent+label')
fig1.show()


df_plot = df.dropna(subset=['price', 'rating'])
fig2 = px.scatter(df_plot, x="rating", y="price",
                 hover_data=['hotel_name'],
                 color="price",
                 title="العلاقة بين تقييم الفندق وسعره")
fig2.show()
```
